Deskripsi Tugas
Pada tugas pratikum ini Anda akan menggunakan data "Wisconsin Breast Cancer". Data tersebut terdiri dari 569 data yang digunakan untuk mendiagnonis jenis kanker Malignant (M) dan Benign (B). Tugas Anda adalah,

1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

2. Lakukan proses encoding pada kolom "diagnosis".

3. Lakukan proses standardisasi pada semua kolom yang memiliki nilai numerik.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

Langkah 1 ‚Äì Memuat Data

```python
# Ganti dengan path file dataset yang kamu punya
df = pd.read_csv("/content/drive/MyDrive/Machine Learning/wbc.csv")

# Tampilkan 5 baris pertama
df.head()
```

Langkah 2 ‚Äì Pisahkan Variabel yang Digunakan

Dataset biasanya memiliki kolom ID atau kolom lain yang tidak relevan.
Pada dataset Wisconsin Breast Cancer, kolom id tidak diperlukan.

```python
# Hapus kolom yang tidak digunakan (misalnya 'id' dan jika ada kolom 'Unnamed: 32')
df = df.drop(columns=['id'], errors='ignore')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Cek lagi struktur data
df.info()
```

Langkah 3 ‚Äì Encoding pada Kolom Diagnosis

Kolom diagnosis berisi label M (Malignant) dan B (Benign).
Kita lakukan encoding menjadi angka:

*   M ‚Üí 1
*   B ‚Üí 0





```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
# Cek hasil setelah encoding
df[['diagnosis']]
```

```python
# Hitung kembali distribusi nilai
df['diagnosis'].value_counts()
```

Langkah 4 ‚Äì Standardisasi pada Kolom Numerik

Semua kolom numerik (selain diagnosis) perlu distandarisasi agar memiliki skala yang seragam.

```python
# Target (hasil diagnosis)
y = df['diagnosis']

# Fitur numerik (selain diagnosis)
X = df.drop(columns=['diagnosis'])
```

```python
scaler = StandardScaler()

# Transformasi ke bentuk array
X_scaled = scaler.fit_transform(X)

# Ubah kembali ke DataFrame agar lebih mudah dibaca
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
df_final = pd.concat([y, X_scaled], axis=1)

# Tampilkan 10 data teratas setelah standardisasi
df_final.head(10)
```
